```python
from pyscbwrapper import SCB
```

```python
#Fetch and print out the specific table "HushallT33DesoN" found in SCBs tree like API structure
scb = SCB('sv', 'HE', 'HE0111', 'HE0111YDeSo', 'HushallT33DesoN')
scb.info()
```

```python
#Get and print out the variables of the table
scb.get_variables()
```

```python
import re
#Use regular expression to filter each object which starts with "0180C" (Each DeSO area of Stockholm municipality starts with this)
regioner = scb.get_variables()['region']
r = re.compile('^0180C')
deso = list(filter(r.match, regioner))

#Create a query of the table with specified parameters for region, upplåtelseform, tabellinnehåll and år
scb.set_query(region=deso,
              upplåtelseform=["äganderätt",
                            "bostadsrätt",
                            "hyresrätt",
                            "uppgift saknas",
                            "totalt"],
              tabellinnehåll=["Antal personer"],
              år=["2020"])
```

```python
#Extract the table from the API using the query created above
scb_data = scb.get_data()
```

```python
#Print out the queried table
scb_data
```

```python
#Extract the data from the table
scb_uttag = scb_data['data']
scb_uttag
```

```python
#Extract the deso key values into a list
koder = scb.get_query()['query'][0]['selection']['values']
#Create a dictionary of the deso key values which will be used in the next step
desodic = {}
for i in range(len(koder)):
  desodic[koder[i]] = koder[i]
desodic
```

```python
#Create a new dictionary called desodata for which the data will be reformatted to a desired structure
#Each deso key value will be represented as a key in the dictionary and all the values will be a nested dictionary with key, value pairs
desodata = {}
for kod in desodic:
    desodata[desodic[kod]] = {}
    for i in range(len(scb_uttag)):
        if scb_uttag[i]['key'][0] == kod:
            desodata[desodic[kod]][scb_uttag[i]['key'][1]] = scb_uttag[i]['values']
desodata
```

```python
import pandas as pd
#Create a pandas dataframe from the dictionary "desodata"
df1 = pd.DataFrame.from_dict(desodata, orient='index').rename_axis('deso').reset_index()
#Rename column 'ÄG/ANDEL' to 'ÄG_ANDEL to enable pandas manipulation on the column at a later stage
df1.rename(columns={'ÄG/ANDEL':'ÄG_ANDEL'}, inplace=True)
df1
```

```python
#Explode the values (which are nested in lists) into new columns and drop the old columns
df1[['ownership']] = pd.DataFrame(df1.ÄG_ANDEL.tolist(), index= df1.index)
df1[['tenant_owned']] = pd.DataFrame(df1.BOSTADSRÄTT.tolist(), index= df1.index)
df1[['tenancy']] = pd.DataFrame(df1.HYRESRÄTT.tolist(), index= df1.index)
df1[['other']] = pd.DataFrame(df1.ÖVRIGT.tolist(), index= df1.index)
df1[['tot']] = pd.DataFrame(df1.TOTALT.tolist(), index= df1.index)
df1.drop(['ÄG_ANDEL', 'BOSTADSRÄTT', 'HYRESRÄTT', 'ÖVRIGT', 'TOTALT'], axis=1, inplace=True)
df1
```

```python
#Export the processed table into a .csv file
df1.to_csv('../csv-output/deso_stockholm_boende.csv')
```
